```python
import random as rnd

age = {0:'<18', 1:'>=18'}
bachelor_degree = {0:'Dont have', 1:'Have'}
gender = {0:'Female', 1:'Male'}
credit = {0:'Dont have', 1:'Have'}
is_gamer = {0: 'No', 1:'Yes'}
data = []
for i in range(0, 333):
    row = []
    row.append(round(rnd.uniform(100,5000)))
    row.append(age[rnd.randint(0, 1)])
    row.append(bachelor_degree[rnd.randint(0, 1)])
    row.append(gender[rnd.randint(0, 1)])
    row.append(credit[rnd.randint(0, 1)])
    row.append(is_gamer[rnd.randint(0, 1)])
    data.append(row)
df = pd.DataFrame(data, columns=['salary','age', 'degree', 'gender', 'credits', 'is_gamer'])
```

```python
df.replace({'<18': 0, '>=18':1, 'have':1, 'dont have':0, 'Male': 1, 'Female':0, 'Have':1, 'Dont have':0, 'Yes':1, 'No':0}, None, True)

df.to_csv('owner.csv')
```
